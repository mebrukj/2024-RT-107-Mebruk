```python
def unique_letters(string):
    reviewed = set(string)
    for letter in reviewed:
        yield letter

for letter in unique_letters("hello"):
    print(letter, end=' ')
```

```python
def unique_string(string):
    seen = set(string)
    for letter in seen:
        yield letter

for letter in unique_letters('hello'):
    print(letter, end=' ')
```
