```python
import numpy as np
import pandas as pd
%matplotlib inline
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
import folium
```

```python
import io

URL = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-DV0101EN-SkillsNetwork/Data%20Files/historical_automobile_sales.csv"
df = pd.read_csv(URL)
print('Data downloaded and read into a dataframe!')

df.columns
```

```python
#1.1
df_lineplot = df.groupby('Year')['Automobile_Sales'].mean()
df_lineplot.plot(kind='line', figsize=(10,6))
plt.xlabel('Year')
plt.ylabel('Automobile Sales')
plt.title('Average Auto Sales per Month')

##df_lineplot.head(30)
```

```python
#1.2 
df_vehicles = df.groupby(['Year','Vehicle_Type'], as_index=False)['Automobile_Sales'].mean()
df_vehicles.set_index('Year', inplace=True)
df_vehicles = df_vehicles.groupby(['Vehicle_Type'])['Automobile_Sales']
df_vehicles.plot(kind='line', figsize=(25,20))
plt.xlabel('Year')
plt.ylabel('Avg Automobile Sales')
plt.title('Automobile Sales by Vehicle Type')
annotations = [
    (1980, 640, "Recession Period 1"),
    (1981, 630, "Recession Period 2"),
    (1991, 25, "Recession Period 3"),
    (2000, 30, "Recession Period 4"),
    (2008, 40, "Great Recession"),
    (2020, 40, "Covid-19 Recession")
]

for (x_coord, y_coord, text) in annotations:
    plt.text(x_coord, y_coord, text, fontsize=12, ha='right')

plt.legend()
plt.show()
```
